```python
def mon(data):
    x = []
    m = []
    for i in xrange(1,13):
        x = data[data.month == i].index
        x = data.ix[x,"year":"month"]
        m.append(len(x))
    return m

month_11 = mon(ix_11)
month_12 = mon(ix_12)
month_13 = mon(ix_13)
month_14 = mon(ix_14)
month_15 = mon(ix_15)
print month_15
```

```python
def cum_mon(data):
    x = 0
    l = []
    for i in data:
        x += i
        l.append(x)
    return np.array(l)

mon_11 = cum_mon(month_11)
mon_12 = cum_mon(month_12) + max(mon_11)
mon_13 = cum_mon(month_13) + max(mon_12) 
mon_14 = cum_mon(month_14) + max(mon_13)
mon_15 = cum_mon(month_15) + max(mon_14)
print mon_11
print mon_12
print mon_13
print mon_14
print mon_15

print df_train.shape

#plt.plot(,df_train[month_11,"price_doc"])
#plt.xlabel(
```

```python
mon_list = np.hstack((mon_11,mon_12,mon_13,mon_14,mon_15))
mon_list =  mon_list[7:54]
print mon_list
```

```python
mean_list = []
median_list = []
for i in xrange(len(mon_list)):
    if i == 0:
        mean_list.append(int(df_train.price_doc[0:mon_list[i]].mean()))
        median_list.append(int(df_train.price_doc[0:mon_list[i]].median()))
    else:
        mean_list.append(int(df_train.price_doc[mon_list[i-1]:mon_list[i]].mean()))
        median_list.append(int(df_train.price_doc[mon_list[i-1]:mon_list[i]].median()))

print len(mean_list)
```

```python
plt.hold(True)
plt.xlabel("month")
plt.ylabel("price_doc")
plt.plot(xrange(len(mean_list)),mean_list,"b--")
plt.plot(xrange(len(median_list)),median_list,"r--")
plt.show()
```

```python
mm_list = (np.array(median_list) + np.array(mean_list))/2

growth = []
for i in xrange(len(median_list)-1):
    x = np.round(float((median_list[i+1]-median_list[i]))/median_list[i] * 100.)
    growth.append(x)
print np.array(growth)
print len(growth)


# m_growth is growth rate[%]
m_growth = []
for i in xrange(len(mm_list)-1):
    x = np.round(float((mm_list[i+1]-mm_list[i]))/mm_list[i] * 100.)
    m_growth.append(x)
print np.array(m_growth)
# m_growth = %
```

```python
plt.hold(True)
plt.plot(xrange(len(m_growth)),m_growth,"b-")
plt.plot(xrange(len(growth)),growth,"r-")
plt.show()
```
